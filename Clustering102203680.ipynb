#K Means Clustering Method

```python
from pycaret.clustering import *
import pandas as pd
from prettytable import PrettyTable

pt = PrettyTable()
myDataSet = pd.read_csv("/content/Sample - Superstore.csv", encoding='ISO-8859-1', nrows=100)

kMeanClusteringParameters = setup(data=myDataSet, verbose=False)

num_clusters = [3, 4, 5]
svalues = []
cvalues = []
dvalues = []

def get_parameters(num_clusters):
    model = create_model('kmeans', num_clusters=num_clusters)
    prediction = assign_model(model)
    performance = pull()
    svalues.append(performance['Silhouette'][0])
    cvalues.append(performance['Calinski-Harabasz'][0])
    dvalues.append(performance['Davies-Bouldin'][0])

def get_ans():
  for num in num_clusters:
    get_parameters(num)

print("For no data processing")
get_ans()
print("For using normalisation")
normparam = setup(data=myDataSet,normalize = True, normalize_method = 'zscore', verbose=False)
get_ans()
print("For using transform")
transformparam=setup(data=myDataSet,transformation = True, transformation_method = 'yeo-johnson', verbose=False)
get_ans()
print("For using PCA")
pcaparam=setup(data=myDataSet,transformation = True,pca = True, pca_method = 'linear', verbose=False)
get_ans()
print("For using N+T")
ntparam=setup(data=myDataSet,transformation = True, normalize = True,normalize_method = 'zscore',
      transformation_method = 'yeo-johnson', verbose=False)
get_ans()
print("For using N+T+PCA")
ntpcaparam=setup(data=myDataSet,transformation = True, normalize = True, pca = True,
      normalize_method = 'zscore', transformation_method = 'yeo-johnson',
      pca_method = 'linear',  verbose=False)
get_ans()
```

```python
pt.field_names=["Parameters","No data preprocessing c=3","No data preprocessing c=4 ","No data preprocessing c=5","Using Normalisation c=3","Using Normalisation c=4","Using Normalisation c=5","Using Transform c=3","Using Transform c=4","Using Transform c=5","Using PCA c=3","Using PCA c=4","Using PCA c=5","Using N+T c=3","Using N+T c=4","Using N+T c=5","Using N+T+PCA c=3","Using N+T+PCA c=4","Using N+T+PCA c=5"]
pt.add_row(["Silhouette",*svalues])
pt.add_row(["Calinski-Harabasz",*cvalues])
pt.add_row(["Davies-Bouldin",*dvalues])
print("USING K-MEANS CLUSTERING")
print(pt)
```

```python

 # To save output as .csv file
import csv

with open("kmeans_output.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(pt.field_names)
    for row in pt._rows:
        writer.writerow(row)
```
